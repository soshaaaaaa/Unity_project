TF-IDF

```python
import math
doc1 = ["new","york","times"]
doc2 = ["new","york","post","new"]
doc3 = ["los","angeles","times","post","new","post"]

def compl_TF(word_list):
    tf = {}
    dl_words = len(word_list)
    for i in word_list:
        tf[i] = tf.get(i,0) + 1 / dl_words
    return tf
print("TF для doc1:", compl_TF(doc1))
print("TF для doc2:", compl_TF(doc2))
print("TF для doc3:", compl_TF(doc3))

def count_word(docs):
    word_podschet = {}
    for i in docs:
        un_words = set(i)
        for j in un_words:  
            if j in word_podschet:
                word_podschet[j] += 1
            else:
                word_podschet[j] = 1
    return word_podschet
word_doc_counts = count_word([doc1, doc2, doc3])
print("Сколько документов содержит каждое слово:")
print(word_doc_counts)

def compl_IDF(word_counts, total_docs):
    idf = {}
    for word, count in word_counts.items():
        idf[word] = math.log10(total_docs / count)
    return idf
total_documents = 3
idf_scores = compl_IDF(word_doc_counts, total_documents)
print(idf_scores)

def compl_tfidf(tf_dict, idf_dict):
    tfidf = {}
    for word, tf_value in tf_dict.items():
        tfidf[word] = tf_value * idf_dict[word]
    return tfidf

tfidf1 = compl_tfidf(compl_TF(doc1), idf_scores)
tfidf2 = compl_tfidf(compl_TF(doc2), idf_scores)
tfidf3 = compl_tfidf(compl_TF(doc3), idf_scores)

print("doc1:", tfidf1)
print("doc2:", tfidf2)
print("doc3:", tfidf3)

def cosine_sim_tfidf(tfidf1, tfidf2):
    ob_words = set(tfidf1.keys()) & set(tfidf2.keys())
    numer = sum(tfidf1[word] * tfidf2[word] for word in ob_words)
    norm1 = math.sqrt(sum(val**2 for val in tfidf1.values()))
    norm2 = math.sqrt(sum(val**2 for val in tfidf2.values()))
    if norm1 == 0 or norm2 == 0:
        return 0
    
    return numer / (norm1 * norm2)
print("doc1-doc2:", round(cosine_sim_tfidf(tfidf1, tfidf2), 3))
print("doc1-doc3:", round(cosine_sim_tfidf(tfidf1, tfidf3), 3))
print("doc2-doc3:", round(cosine_sim_tfidf(tfidf2, tfidf3), 3))
```
